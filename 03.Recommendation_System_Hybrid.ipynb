## Building a Hybrid Recommendation System

```python
import os
import shutil

import kaggle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
if "tmdb_5000_credits.csv" or "tmdb_5000_movies.csv" not in os.listdir("datasets/"):
    kaggle.api.dataset_download_files("tmdb/tmdb-movie-metadata")
    shutil.unpack_archive("tmdb-movie-metadata.zip", extract_dir="./datasets/")
```

```python
credits = pd.read_csv("datasets/tmdb_5000_credits.csv")
```

```python
movies_df = pd.read_csv("datasets/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies_df.head()
```

## EDA

```python
print("Credits:", credits.shape)
print("Movies Dataframe:", movies_df.shape)
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})
movies_df_merge = movies_df.merge(credits_column_renamed, on="id")
movies_df_merge.head()
```

```python
movies_cleaned_df = movies_df_merge.drop(
    columns=["homepage", "title_x", "title_y", "status", "production_countries"]
)
movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

### Using Weighted average for each movie's  Average Rating



```python
# Calculate all the components based on the above formula
v = movies_cleaned_df["vote_count"]
R = movies_cleaned_df["vote_average"]
C = movies_cleaned_df["vote_average"].mean()
m = movies_cleaned_df["vote_count"].quantile(0.70)
```

```python
movies_cleaned_df["weighted_average"] = ((R * v) + (C * m)) / (v + m)
```

```python
movies_cleaned_df["weighted_average"]
```

```python
movie_sorted_ranking = movies_cleaned_df.sort_values(
    "weighted_average", ascending=False
)
movie_sorted_ranking[
    ["original_title", "vote_count", "vote_average", "weighted_average", "popularity"]
].head(20)
```

```python
weight_average = movie_sorted_ranking.sort_values("weighted_average", ascending=False)
plt.figure(figsize=(12, 5))
axis1 = sns.barplot(
    x=weight_average["weighted_average"].head(10),
    y=weight_average["original_title"].head(10),
    data=weight_average,
)
plt.xlim(4, 10)
plt.title("Best Movies by average votes", weight="bold")
plt.xlabel("Weighted Average Score", weight="bold")
plt.ylabel("Movie Title", weight="bold");
```

```python
popularity = movie_sorted_ranking.sort_values("popularity", ascending=False)
plt.figure(figsize=(10, 5))
ax = sns.barplot(
    x=popularity["popularity"].head(10),
    y=popularity["original_title"].head(10),
    data=popularity,
)

plt.title("Most Popular by Votes", weight="bold")
plt.xlabel("Score of Popularity", weight="bold")
plt.ylabel("Movie Title", weight="bold")
```

### Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(
    movies_cleaned_df[["weighted_average", "popularity"]]
)
movie_normalized_df = pd.DataFrame(
    movie_scaled_df, columns=["weighted_average", "popularity"]
)
movie_normalized_df.head()
```

```python
movies_cleaned_df[
    ["normalized_weight_average", "normalized_popularity"]
] = movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df["score"] = (
    movies_cleaned_df["normalized_weight_average"] * 0.5
    + movies_cleaned_df["normalized_popularity"] * 0.5
)
movies_scored_df = movies_cleaned_df.sort_values(["score"], ascending=False)
movies_scored_df[
    ["original_title", "normalized_weight_average", "normalized_popularity", "score"]
].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values("score", ascending=False)

plt.figure(figsize=(10, 5))

ax = sns.barplot(
    x=scored_df["score"].head(10),
    hue=scored_df["original_title"].head(10),
    data=scored_df,
    palette="deep",
)

# plt.xlim(3.55, 5.25)
plt.title("Best Rated & Most Popular Blend", weight="bold")
plt.xlabel("Score", weight="bold")
plt.ylabel("Movie Title", weight="bold")
```
